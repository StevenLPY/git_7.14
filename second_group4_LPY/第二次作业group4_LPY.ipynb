```python
import pandas as pd
import numpy as np
import talib as ta
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
data=pd.read_excel('sz50.xlsx',sheetname=None, index_col='datetime')
```

```python
print(data.keys())
```

```python
stock = pd.read_excel('sz50.xlsx',sheetname='600036.XSHG', index_col='datetime')
```

```python
print(type(np.array(stock)))
```

```python
print(ta.MA(stock.close.values, 10)[-5:])
```

```python
MA=ta.MA(stock.close.values, 10)
ndArray=np.array(MA)
```

```python
MA_Series=pd.Series(ndArray)
MA_Series.index=data['600036.XSHG'].index
close_series=pd.Series(stock.close)
close_series.index=data['600036.XSHG'].index
plt.figure(figsize=(15, 7))
plt.plot(close_series)
plt.plot(MA_Series)
plt.show()
```

```python
plt.figure(figsize=(15, 7))
i=0
for name in data.keys():
    closeArray=data[name]['close'].values
    ROCR100=ta.ROCR100(closeArray,5)
    ndArray=np.array(ROCR100)
    ROCR100S=pd.Series(ndArray)
    ROCR100S.index=data[name].index
    plt.plot(ROCR100S)
    i+=1
    if i>=5:
        break
plt.show()
    
```
